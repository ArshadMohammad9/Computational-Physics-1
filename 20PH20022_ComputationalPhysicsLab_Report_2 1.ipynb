```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#part a
def function(A,x_0):
    num_iter=50
    x=np.zeros(num_iter)
    x[0]=x_0
    for i in range (num_iter-1):
        x[i+1]=A*x[i]*(1-x[i])
    
    x_axis=np.arange(num_iter)
    plt.plot(x_axis,x)
    plt.plot(x_axis, x_axis*0 + (A-1)/A )

plt.figure()
function(0.55,0.56)

plt.figure()
function(1.5,0.56)

plt.figure()
function(2.5,0.56)

plt.figure()
function(3.2,0.56)
```

```python
#part c
n = 3000
A = 3.0
x0 = 0.980
x1 = x0 + 0.01

X0 = np.array([])
X1 = np.array([])

for i in range (n):
        x0 = A*x0*(1-x0)
        X0 = np.append(X0, x0)
        x1 = A*x1*(1-x1)
        X1 = np.append(X1, x1)

        
plt.figure()
plt.plot(np.log(abs(X0-X1)/0.01))
plt.figure()
plt.loglog(np.log(abs(X0 - X1) / 0.01))
plt.show()
```
