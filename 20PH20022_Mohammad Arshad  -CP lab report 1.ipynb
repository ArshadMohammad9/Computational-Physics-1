```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#double precission floating point number
x=np.float64(1.0)
print(x.dtype)
x=1
y=1.0+x
n=0
x_axis=[]
y_axis=[]
for z in range (1,100):
    if(y>1):
        x=x/2
        y=x+1
        x_axis.append(z)
        y_axis.append(y)
        n=n+1
print("Smallest number in the machine is",2*x)
print(n-1)

plt.figure()
plt.plot(x_axis,y_axis)

plt.figure()
plt.plot(x_axis,y_axis)
plt.xscale('log')
plt.yscale('log')
```

```python
#single precission floating point number
x=np.float32(1.0)
print(x.dtype)
y=x+np.float32(1.0)
n=0
x_axis=[]
y_axis=[]
for z in range (1,100):
    if(y>1.0):
        x=np.float32(0.5)*x
        y=x+np.float32(1.0)
        x_axis.append(z)
        y_axis.append(y)
        n=n+1
print("Smallest number in the machine is",np.float32(2.0)*x)
print(n-1)

plt.figure()
plt.plot(x_axis,y_axis)

plt.figure()
plt.plot(x_axis,y_axis)
plt.xscale('log')
plt.yscale('log')
```

```python
a=10
fraction_deveation=[]
v=[]
b=[]
c=[]
for i in range(25):
    x=a**(-i)
    fraction_deveation.append(x)
    y=(x**2 +1)**(0.5) - 1
    b.append(y)
    w=x**2/((x**2 + 1)**(0.5) + 1)
    c.append(w)
    A= np.abs((y-w)/w)
    v.append(A)

    
print(b)
plt.figure()
plt.plot(fraction_deveation,b)

print("\n")
print(c)
plt.figure()
plt.plot(fraction_deveation,c)

print("\n")
print(v)
plt.figure()
plt.plot(fraction_deveation,v)
plt.xscale('log')
plt.yscale('log')
    
```
